```python
import networkx as nx
import matplotlib.pyplot as plt
import algorithmx
import pandas as pd

G = nx.gnp_random_graph(15, 0.3, 135)
# total de vertices e arestas
n_vertices = G.number_of_nodes()
n_arestas = G.number_of_edges()

print('vertices: ', n_vertices, '\narestas: ', n_arestas)
pares = G.edges()
no_1 = []
no_2 = []
for u, v in pares:
    no_1.append(u)
    no_2.append(v)

print(pares)
#for i in range(len(no_1)):
#    print('par', i,':(',no_1[i],',',no_2[i],')')

nx.draw_circular(G, with_labels=True)
nx.write_edgelist(
    G,
    "grafoBase.csv",
    delimiter = ",",
    data = False,
    encoding ='utf-8'
)


```

```python


with open("grafoBase.csv") as f:
    arestas = f.read().splitlines()
node_1 = []
node_2 = []

for i, cont in zip(arestas, range(len(arestas))):
    if cont == 0:
         continue
    node_1.append(i.split(',')[0])
    node_2.append(i.split(',')[1])
    

df = pd.DataFrame({'node_1': node_1, 'node_2': node_2})

print(df)
```

```python
G_l = nx.from_pandas_edgelist(
    df,
    "node_1",
    "node_2",
    create_using=nx.Graph())

nx.draw_circular(G_l, with_labels= True)
plt.show()
# teste para verificar vizinhos
G2 = nx.read_edgelist("grafoBase.csv", delimiter=",")
```
